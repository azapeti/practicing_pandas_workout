```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
df = pd.read_json('../../../data/cities.json')
```

```python
#for digit grouping and control floating-point numbers

pd.set_option('display.float_format', '{:,.2f}'.format)

df['population'] = df['population'].astype('float64')
```

### Task_1
#### What are the mean and median populations for these 1,000 largest cities? What does that tell you?

```python
print(df['population'].describe())

# mean is ~2x of the median so there are some cities with real high population (outliers)
```

### Task_2
#### Along these lines, if you remove the 50 most populous cities, what happens to the mean population? What happens to the median?

```python
cleaned_df = df.drop(df.sort_values(by= 'population').tail(50).index)#.reset_index(drop= True)
cleaned_df
```

```python
cleaned_df['population'].describe()
# now the mean in only  1.3x of the median
```

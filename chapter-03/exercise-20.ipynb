### Beyond the exercise_3
#### Find the city or cities with latitudes more than two standard deviations from the mean.

```python
mean_lat = df['latitude'].mean()
std_lat = df['latitude'].std()

df.loc[
    (
        (df['latitude'] > (mean_lat + 2 * std_lat)) |
        (df['latitude'] < (mean_lat - 2 * std_lat))
    ),
    ['city', 'latitude']
]
```

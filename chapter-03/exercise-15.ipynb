```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
df = (
    pd.read_csv('../../../data/nyc_taxi_2019-01.csv',
    sep= ',',
    header= 0,
    usecols = ['passenger_count', 'trip_distance', 'total_amount', 'payment_type'])
)
```

```python
df.memory_usage()
```

#### How many taxi rides had more than eight passengers?

```python
len(df.query('passenger_count > 8'))
```

```python
len(df[df['passenger_count'] > 8])
```

```python
import time

# Method 1: Using df.query()
start_time = time.perf_counter()
result1 = len(df.query('passenger_count > 8'))
query_time = time.perf_counter() - start_time

# Method 2: Using boolean indexing
start_time = time.perf_counter()
result2 = len(df[df['passenger_count'] > 8])
indexing_time = time.perf_counter() - start_time

print(f"Query method time: {query_time:.6f} seconds, Result: {result1}")
print(f"Boolean indexing time: {indexing_time:.6f} seconds, Result: {result2}")
```

#### How many rides traveled a below-average distance but cost an above-average amount?

```python
mean_distance = df['trip_distance'].mean()
mean_cost = df['total_amount'].mean()

len(df.query('trip_distance < @mean_distance & total_amount > @mean_cost'))
```

```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
df = pd.read_csv(
    '../../../data/nyc_taxi_2020-01.csv',
    header= 0,
    sep= ',',
    usecols= ['passenger_count', 'total_amount', 'payment_type']
)
```

```python
df = pd.read_csv(
    '../../../data/nyc_taxi_2020-01_na_dropped.csv',
    header= 0,
    sep= ',',
    usecols= ['passenger_count', 'total_amount', 'payment_type'],
    dtype= {
        'passenger_count': np.int8,
        'total_amount': np.float32,
        'payment_type': np.int8
    }
)
```

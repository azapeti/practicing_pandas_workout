```python
df_2 = pd.read_csv(
    '../../../data/nyc_taxi_2020-01.csv',
    header= 0,
    sep= ',',
    usecols= ['VendorID', 'trip_distance', 'tip_amount', 'total_amount']
)
```

```python
df_2.info(verbose=True, memory_usage=True, show_counts=True)
```

```python
# ~65,000 VendorID values are missing; not crucial, so filled with 0.

df_2_filled = df_2.fillna(0)
```

```python
# I asked an LLM to provide the minimum and maximum values for NumPy float and int types.

# import numpy as np

# Floating-point types
float_types = [np.float16, np.float32, np.float64]

# Integer types (signed and unsigned)
int_types = [
    np.int8, np.int16, np.int32, np.int64,
    np.uint8, np.uint16, np.uint32, np.uint64
]

print("=== Floating-Point Types ===")
for ftype in float_types:
    finfo = np.finfo(ftype)
    print(f"Type: {ftype.__name__}")
    print(f"Min negative value: {-finfo.max:.50f}")
    print(f"Max negative value: {-finfo.min:.50f}")
    print(f"Min positive value: {finfo.min:.50f}")
    print(f"Max positive value: {finfo.max:.50f}")
    print("---")

print("\n=== Integer Types ===")
for itype in int_types:
    iinfo = np.iinfo(itype)
    print(f"Type: {itype.__name__}")
    if "uint" in itype.__name__:  # Unsigned integers
        print(f"Min value: {iinfo.min}")
        print(f"Max value: {iinfo.max}")
    else:  # Signed integers
        print(f"Min value: {iinfo.min}")
        print(f"Max value: {iinfo.max}")
    print("---")
```

```python
df_2_filled.min(), df_2_filled.max()
```

```python
df_2_filled['VendorID'] = df_2_filled['VendorID'].astype(np.int8)  # only integers, and values are between -128 and 127
df_2_filled['trip_distance'] = df_2_filled['trip_distance'].astype(np.float32)  # floating-point numbers, and the maximum value is higher than 65504
df_2_filled['tip_amount'] = df_2_filled['tip_amount'].astype(np.float16)  # min and max values are within ±65504
df_2_filled['total_amount'] = df_2_filled['total_amount'].astype(np.float16)  # min and max values are within ±65504
```

```python
df_2_filled.info()
# after the dtype conversion, the df consumes 140.5 MB less memory.
```

##### Instead of removing `NaN` values from the `VendorID` column, set it to a new value: How does that affect your specifications and cleaning of the data?

```python
# Since in the first step I did not choose to drop the rows with NaN values but instead filled the VendorID,
# I will now use dropna to remove the rows that contain NaN values.

df_3 = pd.read_csv(
    '../../../data/nyc_taxi_2020-01.csv',
    header= 0,
    sep= ',',
    usecols= ['VendorID', 'trip_distance', 'tip_amount', 'total_amount']
)

df_3_dropped = df_3.dropna()

df_3_dropped.info()

# dropping the ~65,000 rows that contained NaN values consumed only about 2 MB more memory
# than if I had left them in the dataset
```

```python
df_3_dropped['VendorID'] = df_3_dropped['VendorID'].astype(np.int8)  # only integers, and values are between -128 and 127
df_3_dropped['trip_distance'] = df_3_dropped['trip_distance'].astype(np.float32)  # floating-point numbers, and the maximum value is higher than 65504
df_3_dropped['tip_amount'] = df_3_dropped['tip_amount'].astype(np.float16)  # min and max values are within ±65504
df_3_dropped['total_amount'] = df_3_dropped['total_amount'].astype(np.float16)  # min and max values are within ±65504
```

```python
df_3_dropped.info()

# After the dtype optimization, dropping the NaN rows only saved 0.6 MB.
```

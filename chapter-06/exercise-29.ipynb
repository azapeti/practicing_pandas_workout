```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
df = pd.read_csv(
    '../../../data/nyc_taxi_2019-01.csv',
     engine = 'pyarrow',
    dtype_backend = 'pyarrow',
    usecols = ['passenger_count', 'trip_distance', 'total_amount'] 
)
```

```python
# Using a descending sort, find the average cost of the 20 longest (in distance) taxi
# rides in January 2019.

round(df.sort_values(by='trip_distance', ascending = False).head(20)['total_amount'].mean(), 2)
```

```python
# Using an ascending sort, find the average cost of the 20 longest (in distance) taxi
# rides in January 2019. Are the results any different?

round(df.sort_values(by='trip_distance').tail(20)['total_amount'].mean(), 2)
```

```python
# Sort by ascending passenger count and descending trip distance. (So, start with
# the longest trip with 0 passengers and end with the shortest trip with 9 passengers.)
# What is the average price paid for the top 50 rides?

(
    df
    .sort_values(by = ['passenger_count', 'trip_distance'], ascending = [True, False])
    .iloc[:50]['total_amount'].mean()
)
```

#### Beyond the exercise_2
##### Let’s assume that multipassenger rides are split evenly among the passengers. Given that assumption, in which 10 multipassenger rides did each individual pay the greatest amount?

```python
df['cost_per_person'] = df['total_amount'] / df['passenger_count']
df.loc[df['passenger_count'] > 1].sort_values(by = 'cost_per_person', ascending = False).head(10)
```

#### Beyond the exercise_3
##### In the exercise solution, I showed that we needed to use `iloc` or `head`/`tail` to retrieve the first/last 20 rows because the index was scrambled after our sort operation. But you can pass `ignore_index=True` to `sort_values`: then the resulting data frame has a numeric index starting at 0. Use this option and `loc` to get the mean `total_amount` for the 20 longest trips.

```python
(
    df.sort_values('trip_distance', ascending=False, ignore_index= True)['total_amount'].loc[:19].mean()
)
```

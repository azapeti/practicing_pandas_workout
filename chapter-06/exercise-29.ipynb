```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
df = pd.read_csv(
    '../../../data/nyc_taxi_2019-01.csv',
     engine = 'pyarrow',
    dtype_backend = 'pyarrow',
    usecols = ['passenger_count', 'trip_distance', 'total_amount'] 
)
```

```python
# Using a descending sort, find the average cost of the 20 longest (in distance) taxi
# rides in January 2019.

round(df.sort_values(by='trip_distance', ascending = False).head(20)['total_amount'].mean(), 2)
```

```python
# Using an ascending sort, find the average cost of the 20 longest (in distance) taxi
# rides in January 2019. Are the results any different?

round(df.sort_values(by='trip_distance').tail(20)['total_amount'].mean(), 2)
```

```python
# Sort by ascending passenger count and descending trip distance. (So, start with
# the longest trip with 0 passengers and end with the shortest trip with 9 passengers.)
# What is the average price paid for the top 50 rides?

(
    df
    .sort_values(by = ['passenger_count', 'trip_distance'], ascending = [True, False])
    .iloc[:50]['total_amount'].mean()
)
```

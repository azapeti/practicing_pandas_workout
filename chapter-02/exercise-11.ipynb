### Beyond the exercise_1
##### In each case, practice using both loc and query:

```python
df['current_net_revenue'] = (df['retail_price'] - df['wholesale_price']) * df['sales']
df
```

```python
# Show the ID and name of products whose net income is in the top 25% quantile.

df.query('current_net_revenue >= current_net_revenue.quantile(0.75)')[['product_id', 'name']]
```

```python
df.loc[df['current_net_revenue'] >= df['current_net_revenue'].quantile(0.75), ['product_id', 'name']]
```

### Beyond the exercise_2

```python
# Show the ID and name of products with lower-than-average sales numbers and
# whose wholesale price is greater than the average.

df.query('(sales < sales.mean()) & (wholesale_price > wholesale_price.mean())')[['product_id', 'name']]
```

```python
df.loc[(df['sales'] < df['sales'].mean()) & (df['wholesale_price'] > df['wholesale_price'].mean()), ['product_id', 'name']]
```

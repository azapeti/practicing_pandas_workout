```python
import pandas as pd
import numpy as np
import random
from pandas import Series, DataFrame
```

```python
"""
For this exercise, I want you to create a data frame that represents a company’s
inventory of five products. Each product has a unique ID number (a two-digit integer
will do), name, wholesale price, retail price, and number of sales in the last month.
You’re making it up, so if you’ve always wanted to be a profitable starship dealer, this is
your chance! Once you have created this data frame, calculate the total net revenue
from all your products.
"""

df = pd.DataFrame({
    'id' : ['23', '96', '97', '15', '87'],
    'name' : ['computer', 'Python Workout', 'Pandas Workout', 'banana', 'sandwich'],
    'wholesale_price' : [500, 35, 35, 0.5, 3],
    'retail_price' : [1000, 75, 75, 1, 5],
    'sales' : [100, 1000, 500, 200, 300]
})


df['total_net_revenue'] = df['sales'] * (df['retail_price']-df['wholesale_price'])

df
```

```python
print(f'Total net revenue: {df['total_net_revenue'].sum()}')
```

### Beyond the exercise

```python
# For what products is the retail price more than twice the wholesale price?

print(df.query('retail_price > (2 * wholesale_price)'))

print('Python Workout and Pandas Workout are the only products with a retail price more than twice its wholesale price.')
```

```python
# How much did the store make from food versus computers versus books? (You
# can retrieve based on the index values, not anything more sophisticated.)

print(f'from food: \n {df.iloc[3:5]['total_net_revenue'].sum()}')
print()
print(f'from computers: \n {df.iloc[0]['total_net_revenue'].sum()}')
print()
print(f'from books: \n {df.iloc[1:3]['total_net_revenue'].sum()}')

```

```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

#### In this exercise, I want you to create the same two-column data frame as in the last exercise. Then, update the values in the temp column so that any value less than 0 is set to 0.

```python
s = pd.read_csv('../../../data/nyc-temps.txt', header = None, names = ['temp']).squeeze()

hours = np.arange(0, 24, 3)

df = pd.DataFrame({
    'temp' : s,
    'hour' : np.tile(hours, len(s) // len(hours) +1)[:len(s)]
})

df.loc[df['temp'] < 0, 'temp'] = 0
```

### Beyond the exercise_1

```python
# Set all the odd temperatures to the mean of all the temperatures.

df = pd.DataFrame({
    'temp' : s,
    'hour' : np.tile(hours, len(s) // len(hours) +1)[:len(s)]
})

df['temp'] = df['temp'].astype(float)

mean_temp = df['temp'].mean()

df.loc[df['temp'] % 2 != 0, 'temp'] = mean_temp

df.head(20)
```

### Beyond the exercise_2

```python
# Set the even temperatures at hours 9 and 18 to 3.

df = pd.DataFrame({
    'temp' : s,
    'hour' : np.tile(hours, len(s) // len(hours) +1)[:len(s)]
})

df.loc[df['temp'] < 0, 'temp'] = 0

df.loc[(df['temp'] % 2 == 0) & (df['hour'].isin([9, 18])), 'temp'] = 3
```

### Beyond the exercise_3

```python
# If the hour is odd, set the temperature to 5.

df = pd.DataFrame({
    'temp' : s,
    'hour' : np.tile(hours, len(s) // len(hours) +1)[:len(s)]
})

df.loc[df['temp'] < 0, 'temp'] = 0

df.loc[(df['hour'] % 2 != 0) , 'temp'] = 5

df
```

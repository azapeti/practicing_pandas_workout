```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
#for digit grouping and control floating-point numbers

pd.set_option('display.float_format', '{:,.2f}'.format)
```

```python
df = pd.DataFrame({
    'id' : ['23', '96', '97', '15', '87'],
    'name' : ['computer', 'Python Workout', 'Pandas Workout', 'banana', 'sandwich'],
    'wholesale_price' : [500, 35, 35, 0.5, 3],
    'retail_price' : [1000, 75, 75, 1, 5],
    'sales' : [100, 1000, 500, 200, 300]
})
```

```python
new_products = pd.DataFrame({
    'id' : ['24', '17', '17'],
    'name' : ['phone', 'apple', 'pear'],
    'wholesale_price' : [200, 0.5, 0.6],
    'retail_price' : [500, 1, 1.2],
    'sales' : [np.nan, np.nan, np.nan]
})
```

```python
df = pd.concat([df, new_products], ignore_index=True)

df
```

```python
df.loc[
    [5, 6, 7],
    ['sales']
] = [52, 200, 143]

df
```

```python
# Finally, recalculate the store’s total net income, including the new products.

df['current_net_revenue'] = (df['retail_price'] - df['wholesale_price']) * df['sales']

df
```

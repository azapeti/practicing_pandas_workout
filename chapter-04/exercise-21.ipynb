```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
df = pd.read_csv(
    '../../../data/nyc-parking-violations-2020.csv',
    header= 0,
    sep= ',',
    usecols= [
        'Date First Observed',
        'Plate ID',
        'Registration State',
        'Issue Date',
        'Vehicle Make',
        'Street Name',
        'Vehicle Color']
)
```

```python
df.head()
```

```python
df.info()
```

```python
# Set the data frame’s index to the Issue Date column.

df = df.set_index('Issue Date')
df
```

```python
# Determine what three makes were most frequently ticketed on January 2, 2020.

df.loc['01/02/2020 12:00:00 AM', 'Vehicle Make'].value_counts().head(3)
```

```python
# Determine the five streets on which cars got the most tickets on June 1, 2020.

df.loc['06/01/2020 12:00:00 AM', 'Street Name'].value_counts().head(5)
```

```python
# Set the index to Vehicle Color.

df = df.reset_index().set_index('Vehicle Color')
df

# NOTE: Reuven called reset_index() before setting the new index, which is very useful.
# In my version, the issue_date column got lost because I didn’t reset the index.
```

```python
# Determine the most common make of vehicles that were either red or blue.

df.loc[['RED', 'BLUE'], ['Vehicle Make']].value_counts().head(1)
```

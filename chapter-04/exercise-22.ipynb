```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
df = pd.read_csv(
    '../../../data/sat-scores.csv',
    sep= ',',
    header= 0,
    usecols= [
        'Year',
        'State.Code',
        'Total.Math',
        'Total.Test-takers',
        'Total.Verbal'
    ]
)

df = df.set_index(['Year', 'State.Code'])
```

```python
df
```

```python
# Determine how many people took the SAT in 2005.

df.loc[2005, 'Total.Test-takers'].sum()
```

```python
# Determine the average SAT math score in 2010 from New York (NY), New Jersey
# (NJ), Massachusetts (MA), and Illinois (IL).

df.loc[(2010, ['NY', 'NJ', 'MA', 'IL']), 'Total.Math'].mean()
```

```python
# Determine the average SAT verbal score in 2012–2015 from Arizona (AZ),
# California (CA), and Texas (TX).

df.loc[(slice(-4, None), ['AZ', 'CA', 'TX']), 'Total.Verbal'].mean()
```

####  Beyond the exercise_3

##### Was the average math score in 2005 higher or lower than that in 2015?

```python
avg_math_2005 = df.loc[2005, 'Total.Math'].mean().round(2)
avg_math_2015 = df.loc[2015, 'Total.Math'].mean().round(2)

if avg_math_2005 > avg_math_2015:
    print(
        f"In 2005, the average math score was ({avg_math_2005}) "
        f"higher than in 2015 ({avg_math_2015})."
    )
else:
    print(
        f"In 2005, the average math score was ({avg_math_2005}) "
        f"lower than in 2015 ({avg_math_2015})."
    )
```

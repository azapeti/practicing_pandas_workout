```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
df = pd.read_csv(
    '../../../data/olympic_athlete_events.csv',
    sep= ',',
    header= 0,
    usecols= ['Age', 'Height', 'Team', 'Year', 'Season', 'City', 'Sport', 'Event', 'Medal']
)

df = df.set_index(['Year', 'Season', 'Sport', 'Event'])
```

```python
df
```

```python
# What is the average age of winning athletes in summer games held between 1936 and 2000?

df = df.sort_index()
df.loc[(slice(1936, 2000), 'Summer'), 'Age'].mean()
```

```python
df.loc[(slice(1936, 2000), 'Summer'), :]
```

```python
# What team has won the most medals in all archery events?

winners_only = df.dropna(subset=['Medal'])

winners_only.loc[slice(None), slice(None), 'Archery']['Team'].value_counts()
```

```python
# Starting in 1980, what is the average height of the “Table Tennis Women’s Team” event?

df.loc[(slice(1980, None), slice(None), slice(None), "Table Tennis Women's Team")]['Height'].mean()
```

```python
# Starting in 1980, what is the average height of
# both “Table Tennis Women’s Team” and “Table Tennis Men’s Team”?

df.loc[(slice(1980, None), slice(None), slice(None), ["Table Tennis Women's Team", "Table Tennis Men's Team"])]['Height'].mean()
```

```python
# How tall was the tallest-ever tennis player in Olympic games from 1980 until 2016?

df.loc[(slice(1980, 2016), slice(None), 'Tennis', slice(None))]['Height'].max()
```

#### Beyond the exercise_1

##### Events occur in either summer or winter Olympic games, but not both. As a result, the "Season" level in our multi-index is often unnecessary. Remove the "Season" level, and then find (again) the height of the tallest tennis player between 1980 and 2016.

```python
df = df.droplevel('Season')
```

```python
from pandas import IndexSlice as idx

df.loc[idx[1980:2016, 'Tennis'], 'Height'].max()
```

#### Beyond the exercise_2

##### In which city were the most gold medals awarded from 1980 onward?

```python
selected_years = df.loc[idx[1980:]]

gold_medals = selected_years[selected_years['Medal'] == 'Gold']['City'].value_counts().head(1)

print(gold_medals)
```

#### Beyond the exercise_3
#### How many gold medals were received by the United States since 1980? (Use the index to select the values.)

```python
df.loc[idx[1980:]].loc[lambda x: (x['Medal'] == 'Gold') & (x['Team'] == 'United States'), 'Medal'].value_counts()
```

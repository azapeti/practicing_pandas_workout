```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
df = pd.read_csv(
    '../../../data/olympic_athlete_events.csv',
    sep= ',',
    header= 0,
    usecols= ['Age', 'Height', 'Team', 'Year', 'Season', 'Sport', 'Medal']
)

# Read in our Olympic data again.
# – Only use these columns: Age, Height, Team, Year, Season, Sport, Medal.
# – Only include games from 1980 to the present.
# – Only include data from these countries: Great Britain, France, United States, Switzerland, China, and India.

df = df.loc[(df['Year'] >= 1980) & (df['Team'].isin(["Great Britain", "France", "United States", "Switzerland", "China", "India"]))]
```

```python
df
```

```python
# What was the average age of Olympic athletes?

df.groupby('Team')['Age'].mean().sort_values()
```

```python
# In which country do players appear to consistently be the youngest?

df.pivot_table(
    index = 'Year',
    columns = 'Team',
    values = 'Age'
).round(2)
```

```python
# How tall were the tallest athletes in each sport in each year?

df.pivot_table(
    index= 'Sport',
    columns= 'Year',
    values= 'Height',
    aggfunc= 'max'
)
```

```python
# How many medals did each country earn each year?

(
    df
    .dropna(subset='Medal')
    .pivot_table(
        index='Team',
        columns='Year',
        values='Medal',
        aggfunc='size'
    )
)
```

#### Beyond the exercise_1
##### Create a pivot table that shows the number of medals each team won per year, with the index including the year and the season in which the games took place.

```python
(
    df
    .dropna(subset= 'Medal')
    .pivot_table(
        index= ['Year', 'Season'],
        columns= 'Team',
        values= 'Medal',
        aggfunc= 'size'
    )
)
```

#### Beyond the exercise_2
##### Create a pivot table that shows both the average age and the average height per year per team.

```python
(
    df.
    pivot_table(
        index= 'Year',
        columns= 'Team',
        values= ['Age', 'Height'],
        aggfunc= 'mean'
    ).round(2)
)
```

#### Beyond the exercise_3
##### Create a pivot table that shows the average age and the average height per year, per team, broken up by year and season.

```python
(
    df.
    pivot_table(
        index= ['Year', 'Season'],
        columns= 'Team',
        values= ['Age', 'Height'],
        aggfunc= 'mean'
    ).round(2)
)
```
